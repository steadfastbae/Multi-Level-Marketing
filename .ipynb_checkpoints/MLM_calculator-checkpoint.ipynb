```python
from sympy.plotting import plot 
from sympy import Symbol
import sympy 
from IPython.display import display

n = Symbol('n')

revenue = 2**n - 1
direct_sale_bonus = (2**n - 2)
team_commission = (2*n-5) * 2**(n-2) + 1


print("revenue")
display(revenue)
print("direct_sale_bonus")
display(direct_sale_bonus)
print("team_commission")
display(team_commission)
```

```python
direct_sale_bonus_rate = 0.2
team_commission_rate = 0.1
profit_revenue_rate = 0.3


direct_sale_bonus = direct_sale_bonus_rate * direct_sale_bonus
team_commission = direct_sale_bonus_rate * team_commission_rate * team_commission
profit = profit_revenue_rate * revenue - direct_sale_bonus - team_commission

print("revenue")
display(revenue)
print("direct_sale_bonus")
display(direct_sale_bonus)
print("team_commission")
display(team_commission)
print("profit")
display(profit)
```

```python
xlabel = "$n_{th}$ Generation"
ylabel = "price"
x_range = (n,0,10)

p1 = plot(
     (profit, x_range),
     xlabel = xlabel, ylabel = ylabel,
     label="profit",
     legend=True,
     show=False)

p2 = plot(
     (revenue, x_range),
     xlabel = xlabel, ylabel = ylabel,
     label="revenue",
     legend=True,
     show=False)

p3 = plot(
     (direct_sale_bonus, x_range),
     xlabel = xlabel, ylabel = ylabel,
     label="direct_sale_bonus",
     legend=True,
     show=False)

p4 = plot(
     (team_commission, x_range),
     xlabel = xlabel, ylabel = ylabel,
     label="team_commission",
     legend=True,
     show=False)



p1.append(p2[0])
p1.append(p3[0])
p1.append(p4[0])
# p1.append(p5[0])


p1.show()
```
